```python
import pycountry
import numpy as np
import pandas as pd

pd.set_option('future.no_silent_downcasting', True)

# Step 1: Clean string versions of NaNs in-place
df_obesity['SpatialDim'] = df_obesity['SpatialDim'].replace(['NAN', 'NaN', 'nan', ''], np.nan)

# Step 2: Prepare country mapping dictionary
country_mapping = {country.alpha_3.upper(): country.name for country in pycountry.countries}

# Step 3: Apply mapping to convert alpha-3 code to full country name
mapped = df_obesity['SpatialDim'].astype(str).str.strip().str.upper().map(country_mapping)

# Step 4: Manual fallback mapping
manual_mapping = {
    "GRL": "Greenland",
    "XKX": "Kosovo",
    "PSE": "Palestine",
    "KOS": "Kosovo",
    'GLOBAL': 'Global',
    'WB_LMI': 'Low & Middle Income',
    'WB_HI': 'High Income',
    'WB_LI': 'Low Income',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'SEAR': 'South-East Asia Region',
    'WPR': 'Western Pacific Region',
    'AMR': 'Americas Region',
    'WB_UMI': 'Upper Middle Income'
}
fallback_mapped = df_obesity['SpatialDim'].astype(str).str.strip().str.upper().map(manual_mapping)

# Step 5: Combine both mappings, assign to SpatialDim (overwrite with full name)
df_obesity['SpatialDim'] = mapped.combine_first(fallback_mapped)

# Step 6: Verify unmapped codes
unmapped_codes = df_obesity[df_obesity['SpatialDim'].isna()]
print("Unmapped 3-letter codes:", unmapped_codes['SpatialDim'].unique())
```

9.Countries where female obesity exceeds male by large margin (same year)

```python
pd.read_sql_query(
   """
SELECT 
    f.Country,
    f.Year,
    f.age_group,
    f.Mean_Estimate AS Female_Obesity,
    m.Mean_Estimate AS Male_Obesity,
    (f.Mean_Estimate - m.Mean_Estimate) AS Difference
FROM df_obesity f
JOIN df_obesity m 
    ON f.Country = m.Country 
    AND f.Year = m.Year 
    AND f.age_group = m.age_group
WHERE f.Gender = 'Female'
  AND m.Gender = 'Male'
  AND (f.Mean_Estimate - m.Mean_Estimate) > 5
ORDER BY Difference DESC
""",myconnection)
```

21.Obesity vs malnutrition comparison by country(any 5 countries)

```python
pd.read_sql_query("""WITH obesity_avg AS (
    SELECT 
        Country,
        AVG(Mean_Estimate) AS Avg_Obesity
    FROM df_obesity
    WHERE Country IN ('Bahamas', 'Iraq', 'Greenland', 'Armenia', 'India')
    GROUP BY Country
),
malnutrition_avg AS (
    SELECT 
        Country,
        AVG(Mean_Estimate) AS Avg_Malnutrition
    FROM df_malnutrition
    WHERE Country IN ('Bahamas', 'Iraq', 'Greenland', 'Armenia', 'India')
    GROUP BY Country
)

SELECT 
    o.Country,
    o.Avg_Obesity,
    m.Avg_Malnutrition
FROM obesity_avg o
JOIN malnutrition_avg m
    ON o.Country = m.Country""",myconnection)
```

```python
pd.read_sql_query("""SELECT
    o.Country,
    AVG(o.Mean_Estimate) AS Obesity_up,
    AVG(m.Mean_Estimate) AS Malnutrition_down
FROM
    df_obesity o
JOIN
    df_malnutrition m ON o.Country = m.Country AND o.Year = m.Year
GROUP BY
    o.Country
HAVING
    AVG(o.Mean_Estimate) > AVG(m.Mean_Estimate)  -- Obesity up, Malnutrition down
ORDER BY
    Obesity_up DESC""",myconnection)
```
